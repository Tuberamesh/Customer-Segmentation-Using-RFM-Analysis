```python

import pandas as pd

df=pd.read_csv("online_retail.csv")


df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df.info()
```

```python
df = df.dropna(subset=['CustomerID'])
df.info()
```

```python
df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]
df.info()
```

```python
df[df['InvoiceNo'].astype(str).str.startswith('C')].shape
```

```python
df = df[df['Quantity'] > 0]

df.info()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
df.info()
```

```python
df['Year'] = df['InvoiceDate'].dt.year
df['Month'] = df['InvoiceDate'].dt.month
df['Day'] = df['InvoiceDate'].dt.day
df['Hour'] = df['InvoiceDate'].dt.hour
df.info()
```

```python
df_clean = df[
    (df['Quantity'] > 0) &
    (df['UnitPrice'] > 0)
].drop_duplicates()

df_clean['Revenue'] = df_clean['Quantity'] * df_clean['UnitPrice']


df_clean.head()
```

```python
snapshot_date = df_clean['InvoiceDate'].max() + pd.Timedelta(days=1)
snapshot_date
```

```python
rfm = df_clean.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (snapshot_date - x.max()).days,  # Recency
    'InvoiceNo': 'nunique',                                    # Frequency
    'Revenue': 'sum'                                           # Monetary
}).reset_index()
```

```python
rfm.columns = ['CustomerID', 'Recency', 'Frequency', 'Monetary']
rfm.head()
```

```python
rfm['R_Score'] = pd.qcut(
    rfm['Recency'],
    5,
    labels=[5, 4, 3, 2, 1]
)
rfm.head()
```

```python
rfm['F_Score'] = pd.qcut(
    rfm['Frequency'].rank(method='first'),
    5,
    labels=[1, 2, 3, 4, 5]
)
rfm.head()
```

```python
rfm['M_Score'] = pd.qcut(
    rfm['Monetary'],
    5,
    labels=[1, 2, 3, 4, 5]
)
rfm.head()
```

```python
rfm[['R_Score','F_Score','M_Score']] = rfm[['R_Score','F_Score','M_Score']].astype(int)
```

```python
rfm['RFM_Score'] = (
    rfm['R_Score'].astype(str) +
    rfm['F_Score'].astype(str) +
    rfm['M_Score'].astype(str)
)
rfm.head()
```

```python
def segment_customer(row):
    if row['R_Score'] >= 4 and row['F_Score'] >= 4:
        return 'Champions'
    elif row['R_Score'] >= 3 and row['F_Score'] >= 3:
        return 'Loyal Customers'
    elif row['R_Score'] >= 4 and row['F_Score'] <= 2:
        return 'New Customers'
    elif row['R_Score'] <= 2 and row['F_Score'] >= 4:
        return 'At Risk'
    else:
        return 'Others'

rfm['Segment'] = rfm.apply(segment_customer, axis=1)

rfm.head()
```

```python
#ðŸ”¥ STEP 6: EDA (Exploratory Data Analysis)
```

```python
import matplotlib.pyplot as plt

rfm['Segment'].value_counts().plot(kind='bar')
plt.title('Customer Segment Distribution')
plt.xlabel('Segment')
plt.ylabel('Number of Customers')
plt.show()
```

```python
rfm.groupby('Segment')['Monetary'].sum().sort_values(ascending=False).plot(kind='bar')
plt.title('Revenue Contribution by Segment')
plt.ylabel('Total Revenue')
plt.show()
```

```python
plt.scatter(rfm['Recency'], rfm['Frequency'])
plt.xlabel('Recency (days)')
plt.ylabel('Frequency')
plt.title('Recency vs Frequency')
plt.show()
```

```python
rfm.to_csv('rfm_customer_segmentation.csv', index=False)
```
